```python
import pandas as pd
import numpy as np
import io
from imblearn.over_sampling import RandomOverSampler
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
# prompt: READ EXCEL FILE
import pandas as pd
# Read the Excel file from Google Drive
excel_file_path = "/content/gdrive/MyDrive/AC/ads_library_campaign_data.xlsx"
df = pd.read_excel(excel_file_path)

# # Print the DataFrame
# print(df)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df[['Primary Text (caption)', 'Category']]
df.dropna(inplace=True)
df.groupby(["Category"])["Primary Text (caption)"].count()
```

```python
df.shape
```

```python
# Convert pandas column to lowercase use map()
df['Category'] = df['Category'].map(str.lower)
print("Convert Pandas column to lower case:\n", df)
```

```python

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['Primary Text (caption)'], df['Category'], test_size=0.20, random_state=12)

# Convert text data into TF-IDF features
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Balance the dataset using oversampling
oversampler = RandomOverSampler()
X_train_resampled, y_train_resampled = oversampler.fit_resample(X_train_tfidf, y_train)

# Initialize and train Support Vector Machine (SVM) classifier
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(X_train_resampled, y_train_resampled)

# Make predictions on the testing set
y_pred = svm_classifier.predict(X_test_tfidf)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import RandomOverSampler
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['Primary Text (caption)'], df['Category'], test_size=0.15, random_state=12)

# Convert text data into TF-IDF features
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Balance the dataset using oversampling
oversampler = RandomOverSampler()
X_train_resampled, y_train_resampled = oversampler.fit_resample(X_train_tfidf, y_train)

# Initialize and train Multinomial Naive Bayes classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_resampled, y_train_resampled)

# Make predictions on the testing set
y_pred = nb_classifier.predict(X_test_tfidf)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Function to predict the category for input text
def predict_category(text):
    # Transform input text into TF-IDF features
    text_tfidf = tfidf_vectorizer.transform([text])

    # Predict the category using the trained model
    predicted_category = svm_classifier.predict(text_tfidf)

    return predicted_category[0]
```
